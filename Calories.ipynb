```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Loading data


```python
cal= pd.read_csv('/content/calories.csv')
```

```python
exercise=pd.read_csv('/content/exercise.csv')
```

```python
#concatenating both table depending upon the user_id
calories=pd.concat([exercise,cal['Calories']],axis=1)
```

```python
calories.replace({"Gender":{'male':0, 'female':1}},inplace=True) #permanently change
```

```python
calories.head()
```

Training and Test data seperation

```python
X=calories.drop(columns=['User_ID','Calories'],axis=1)
Y=calories['Calories']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

XGBoost Regressor

```python
model=XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
#evaluating our data on test data
prediction=model.predict(X_test)
```

```python
print(prediction)
```

```python
#comparing the data with Y_test using Mean Absolute Error
m=metrics.mean_absolute_error(Y_test,prediction)
```

```python
print("Mean Absolute Error =",m)
```
